```python
def get_jgi_geneid(x):
    return x.split(' ')[0]
def get_assembly_geneid(x):
    return x.split(' ')[1]

counts_files=os.listdir('../counts')
for ctr,cf in enumerate(counts_files):
    if cf.endswith('counts.txt') and ctr ==0:
        print(cf)
        cf_header = cf.split('.')[0]
        lib, complexity, ref = cf_header.split('_')
        cur_df = pd.read_csv('../counts/{}'.format(cf),sep='\t',comment='#',header=None)
        cur_df.columns = ['GeneID','Length','Bases','Coverage','Reads','RPKM','Frags','FPKM']
        # Add the ref used
        cur_df.loc[:,'reference'] = ref
        # Add the library name
        cur_df.loc[:,'library'] = lib
        # Add the complexity
        cur_df.loc[:,'complexity'] = complexity
        # Add the community composition
        
        # Add JGI gene id
        cur_df.loc[:,'jgi_gene_id'] = cur_df.apply(lambda x: get_jgi_geneid(x.GeneID),axis=1)
        
        # Add JGI assembly gene id
        cur_df.loc[:,'assembly_gene_id'] = cur_df.apply(lambda x: get_assembly_geneid(x.GeneID),axis=1)
        
        # Add the TPM normalization
        cur_df=cur_df.groupby(['library']).apply(lambda x: x.assign(tpm=lambda x: (x.Reads  / (x.Length/1e3)) / ((x.Reads  / (x.Length/1e3)).sum() / 1e6 )  )).droplevel(0)
```
